```python
import re
extraction_patterns = [
    r'Sector\s?\d+[A-Za-z]?',   # Matches "Sector 10", "Sector 16B"
    r'Gaur City\s?\d*',         # Matches "Gaur City 1", "Gaur City 2"
    r'Noida Extension',
    r'Greater Noida West',
    r'Tech Zone\s?[IV]*',       # Matches "Tech Zone IV"
    r'Knowledge Park\s?[IV]*',
    r'Ecotech\s?[IV]*',
    r'Surajpur',
    r'\bOmicron\s?[IV]*\b',     # Matches "Omicron I", "Omicron II"
    r'\bDelta\s?[IV]*\b',
    r'\bAlpha\s?[IV]*\b',
    r'\bBeta\s?[IV]*\b',
    r'\bGamma\s?[IV]*\b',
    r'\bZeta\s?[IV]*\b',
    r'\bEta\s?[IV]*\b',
    r'\bMu\s?[IV]*\b',
    r'\bChi\s?[IV]*\b',
    r'\bPi\s?[IV]*\b',
    r'\bOmega\s?[IV]*\b',
    r'\bPhi\s?[IV]*\b',
    r'\bSigma\s?[IV]*\b',
    r'\bRho\b',
    r'\bTheta\b',
    r'\bIota\b',
    r'\bKappa\b',
    r'\bLambda\b',
    r'\bNu\b',
    r'\bXi\b',
    r'\bTau\b',
    r'\bUpsilon\b',
    r'\bPsi\b',
    r'Greater Noida'            # Generic fallback
]

# 3. Extraction Function
def extract_locality(title):
    if not isinstance(title, str):
        return None
    
    # Iterate through patterns and return the first match found
    for pattern in extraction_patterns:
        match = re.search(pattern, title, re.IGNORECASE)
        if match:
            return match.group(0).strip()
    return None

# 4. Apply the function to create the new column
print("Extracting localities...")
df['Locality'] = df['full_title'].apply(extract_locality)

# 5. Check results
filled_count = df['Locality'].notnull().sum()
print(f" Successfully filled {filled_count} rows with locality information.")
print(df[['full_title', 'Locality']].head(10))

```

```python
import os
import requests
from typing import Optional


class AzureMapsTools:
    BASE_URL = "https://atlas.microsoft.com/search/fuzzy/json"
    API_VERSION = "1.0"

    def __init__(self):
        self.session = requests.Session()

  
    # Internal Request Handler
   
    def _make_request(self, params: dict) -> dict:
        api_key = os.getenv("AZURE_MAPS_KEY")

        if not api_key:
            raise ValueError("Azure Maps API key not found in environment variables.")

        params.update({
            "api-version": self.API_VERSION,
            "subscription-key": api_key
        })

        response = self.session.get(
            self.BASE_URL,
            params=params,
            timeout=10
        )
        response.raise_for_status()
        return response.json()

    # Get Coordinates
  
    def _get_coordinates(self, society_name: str) -> Optional[tuple]:
        search_query = (
            society_name
            if "Greater Noida" in society_name
            else f"{society_name}, Greater Noida"
        )

        data = self._make_request({
            "query": search_query,
            "countrySet": "IN",
            "limit": 1
        })

        results = data.get("results")
        if not results:
            return None

        pos = results[0]["position"]
        return pos["lat"], pos["lon"]


    # Public Tool Method

    def find_nearby_amenity(
        self,
        society_name: str,
        amenity_type: str,
        radius: int = 5000,
        limit: int = 10
    ) -> str:

        try:
            coords = self._get_coordinates(society_name)
            if not coords:
                return f"Location not found for '{society_name}'."

            lat, lon = coords

            data = self._make_request({
                "query": amenity_type,
                "lat": lat,
                "lon": lon,
                "radius": radius,
                "limit": limit
            })

            results = data.get("results", [])
            if not results:
                return f"No {amenity_type} found within {radius}m of {society_name}."

            filtered = [
                f"- {item.get('poi', {}).get('name', 'Unknown')} "
                f"({int(item.get('dist', 0))}m)"
                for item in results
                if "Sector" not in item.get("poi", {}).get("name", "")
            ]

            return (
                f"Nearby {amenity_type}s near '{society_name}':\n"
                + "\n".join(filtered)
            )

        except requests.RequestException as e:
            return f"API request failed: {str(e)}"
        except Exception as e:
            return f"Unexpected error: {str(e)}"



# Usage
azure_maps = AzureMapsTools()

print("Setup Complete.")
print(azure_maps.find_nearby_amenity("Ace City", "School"))
```

```python
import sqlite3
import pandas as pd

class MarketAnalysisTools:
    def __init__(self, db_path):
        self.db_path = db_path

    def get_market_context(self, locality: str):
        """Fetches property data for a locality to calculate medians."""
        conn = sqlite3.connect(self.db_path)
        # Fetch raw data to calculate the median in pandas (SQL doesn't have a native MEDIAN function)
        query = f"""
            SELECT rent, area_sqft
            FROM nobroker_cleaned 
            WHERE Locality = '{locality}'
        """
        df = pd.read_sql_query(query, conn)
        conn.close()
        
        if df.empty:
            return None
            
        # Calculate price per sqft for every row to find the true middle ground
        df['price_per_sqft'] = df['rent'] / df['area_sqft']
        
        return {
            "locality": locality,
            "median_rent_per_sqft": df['price_per_sqft'].median(),
            "sample_size": len(df)
        }

    def evaluate_deal(self, rent: int, area_sqft: int, locality: str) -> str:
        """Compares a specific property to the localized market median and outputs negotiation advice."""
        market_data = self.get_market_context(locality)
        
        if not market_data:
            return f"No market data available for {locality} to evaluate this deal."
        
        prop_price_per_sqft = rent / area_sqft
        market_median = market_data['median_rent_per_sqft']
        
        # Calculate % difference from market median
        diff = ((prop_price_per_sqft - market_median) / market_median) * 100
        
        # Calculate what the rent SHOULD be based on the median
        fair_rent = market_median * area_sqft
        
        # Helper function to format numbers like 16500 into '16.5k'
        def to_k(amount):
            return f"₹{amount/1000:.1f}k".replace('.0k', 'k')
            
        # Create a realistic negotiation range (e.g., 5% below fair rent to 2% above)
        range_low = to_k(fair_rent * 0.95)
        range_high = to_k(fair_rent * 1.02)
        asking_rent_k = to_k(rent)

        # Format the exact output string for the AI to read
        if diff > 5.0:
            return f"Property is {diff:.1f}% overpriced vs locality median. Recommended negotiation range {range_low}–{range_high}."
        elif diff < -5.0:
            return f"Property is {abs(diff):.1f}% underpriced vs locality median. Excellent deal! Recommend locking it in near the asking price of {asking_rent_k}."
        else:
            return f"Property is priced at Fair Market Value ({abs(diff):.1f}% variance vs median). Recommended negotiation range {range_low}–{range_high}."

# --- Usage Test ---
analysis = MarketAnalysisTools(r"D:\personal Projects\Agentic AI\Reaestate Agent 3\Data\Database\nobroker.db")
print(analysis.evaluate_deal(24000, 1055, "Sector 10"))
```

```python
import os
import uuid
from agno.agent import Agent
from agno.models.azure import AzureOpenAI
from agno.db.sqlite import SqliteDb
from agno.memory import MemoryManager
from rich.pretty import pprint
```

```python
# Importing the SQL tool to allow querying your specific database
from agno.tools.sql import SQLTools

# 1. Define your Database Path
db_path = r"D:\personal Projects\Agentic AI\Reaestate Agent 3\Data\Database\nobroker.db"

# 2. Setup the unified SQLite Database component
db = SqliteDb(db_file=db_path)

# 3. Setup the Central Memory Manager
memory_manager = MemoryManager(db=db)
```

```python
# 4. Initialize the Agent
agent = Agent(
    model=AzureOpenAI(
        id="gpt-4.1-mini-2", # Ensure this matches your Azure deployment ID exactly
        api_version="2024-08-01-preview",
        temperature=0,
    ),
    # Link the SQLite database for session history (Short-Term Memory)
    db=db,
    # Link the Memory Manager for user facts (Long-Term Memory)
    memory_manager=memory_manager,
    
    # Active learning for user preferences
    enable_user_memories=True, 
    
    add_history_to_context=True,
    num_history_runs=3,
    
    # 5. Attach your specific Tools
    tools=[
        azure_maps.find_nearby_amenity,
        analysis.evaluate_deal,
        # Adding the SQL Tool so the agent can query your cleaned housing data
        SQLTools(db_url=f"sqlite:///{db_path}")
    ],
    instructions=[
    "You are a top-tier Real Estate Broker for Greater Noida. You must clearly distinguish between finding 'Housing Inventory', evaluating 'Market Deals', and finding 'Neighborhood Amenities'.",

    "### STEP 0: CLASSIFY USER INTENT",
    "Before calling ANY tool, determine the user's goal:",
    "- **GOAL A: FIND HOUSING** -> User wants flats, apartments, or property options. (Keywords: 'rent', 'flat', 'BHK', 'budget').",
    "- **GOAL B: EVALUATE DEAL** -> User wants to know if a rent price is fair. (Keywords: 'overpriced', 'good deal', 'fair price').",
    "- **GOAL C: FIND AMENITIES** -> User wants nearby facilities. (Keywords: 'school', 'hospital', 'mall', 'market').",

    "### STEP 1: TOOL USAGE MAP (STRICT)",
    "1. **Tool: `SQLTools` (Database)**",
    "   - **Function**: Queries the `nobroker_cleaned` table in the SQLite database.",
    "   - **When to use**: ALWAYS use this first to find flats, rents, or confirm a society's location.",
    "   - **CRITICAL SCHEMA**: `apartment_type` (INTEGER) is used for BHK. `society_name` (TEXT) is the building name. `Locality` (TEXT) is the sector. `rent` (INTEGER). `area_sqft` (INTEGER).",
    
    "2. **Tool: `azure_maps.find_nearby_amenity`**",
    "   - **When to use**: ONLY for GOAL C (Amenities). Finds landmarks on a map.",
    "   - **Input Rules**: `amenity_type` MUST be singular (e.g., 'School', not 'Schools'). `society_name` should be a clean string.",
    
    "3. **Tool: `analysis.evaluate_deal`**",
    "   - **When to use**: ONLY for GOAL B (Evaluate Deal). Compares a property against the sector median and provides a negotiation strategy.",
    "   - **Input Rules**: Requires `rent` (int), `area_sqft` (int), and `locality` (str).",

    "### STEP 2: EXECUTION WORKFLOWS",
    "- **Locality Translation Rule (CRITICAL)**:",
    "   - **Problem**: Users type conversational names like 'chi 5', 'sector pi 2', or 'tech zone 4'.",
    "   - **Database Reality**: The database uses Roman numerals and often omits the word 'sector' (e.g., 'Chi V', 'Pi II', 'Tech Zone IV').",
    "   - **Solution**: You MUST translate Arabic numerals (1, 2, 3, 4, 5) to Roman numerals (I, II, III, IV, V) before querying.",
    "   - **Pattern**: NEVER use strict equality for these. Use wildcards. For 'chi 5', write `WHERE Locality LIKE '%Chi%V%'`. For 'pi 2', write `WHERE Locality LIKE '%Pi%II%'`.",

    "- **SQL Queries ('Noisy Data' Rule)**:",
    "   - **Problem**: Database names can be messy (e.g., 'Godrej Golf Links Crest').",
    "   - **Solution**: NEVER search for exact full strings. Use `LIKE` with split keywords.",
    "   - **Pattern**: `SELECT ... WHERE society_name LIKE '%Godrej%' AND society_name LIKE '%Crest%'`.",
    "   - **BHK Rule**: If the user wants a 2 BHK, ALWAYS write `WHERE apartment_type = 2`.",

    "- **Workflow for Amenities (Goal C)**:",
    "   1. Run SQL to confirm the `society_name` exists.",
    "   2. Call `find_nearby_amenity` using the confirmed society name.",

    "- **Workflow for Market Evaluation (Goal B)**:",
    "   1. Use SQL to find the specific property the user is asking about to get its rent, area, and locality.",
    "   2. Pass those exact numbers into the `evaluate_deal` tool.", 

    "### STEP 3: MEMORY & PERSONALIZATION",
    "- Always check user memories for previously stated preferences (Budget, Locality, BHK).",
    "- Automatically apply remembered constraints to your SQL queries (e.g., if memory says 'Budget is 25000', add `WHERE rent <= 25000`) unless the user explicitly changes it.",

    "### STEP 4: RESPONSE FORMAT",
    "- **Housing**: Use a Markdown Table: `| Title | Society | Locality | BHK | Rent | Area | Link |`",
    "- **Amenities**: Use bullet points with specific names and distances.",
    "- **Deals**: Output the EXACT negotiation string returned by the `evaluate_deal` tool. Format this string in **bold** (e.g., **Property is 12.3% overpriced vs locality median. Recommended negotiation range ₹16k–17k.**)."
],
    # REMOVED: show_tool_calls=True
    markdown=True,
    debug_mode=True
)

```

```python
# --- Interactive Chat Loop ---

# Generate a unique user ID for this specific client
client_id = f"client_{str(uuid.uuid4())[:8]}"

print(f"--- Starting Interactive Session for {client_id} ---")
print("Type 'exit' or 'quit' to end the chat.")
print("Type 'memory' at any time to see what the agent has learned about you.\n")

while True:
    # 1. Capture your input from the Jupyter text box
    user_input = input("User: ")
    
    # Print it to the screen so you have a visual log of what you asked
    print(f"User: {user_input}")
    
    # 2. Handle special commands
    if user_input.lower() in ['exit', 'quit', 'bye']:
        print("\nAgent: Goodbye! Closing the session.")
        
        # Optional: Print the final memory dump before leaving
        print("\n--- Final Central Memory Dump ---")
        memories = agent.memory_manager.get_user_memories(user_id=client_id) if hasattr(agent.memory_manager, 'get_user_memories') else "Check database directly for memories"
        pprint(memories)
        break
        
    if user_input.lower() == 'memory':
        print("\n--- Current Central Memory Dump ---")
        memories = agent.memory_manager.get_user_memories(user_id=client_id) if hasattr(agent.memory_manager, 'get_user_memories') else "Check database directly for memories"
        pprint(memories)
        print("-----------------------------------\n")
        continue

    # 3. Process the input through the Agent
    # We use .run() to ensure Jupyter captures the text properly
    response = agent.run(user_input, user_id=client_id)
    
    # 4. Print the agent's response
    print(f"Agent:\n{response.content}\n")
```
